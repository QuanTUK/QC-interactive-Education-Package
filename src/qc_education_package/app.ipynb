```python
import os
import json
from qc_education_package import InteractiveDCNViewer

try:
    num_qubits = int(os.environ.get("VIEWER_QUBITS", 3))
    initial_state = json.loads(os.environ.get("VIEWER_INITIAL", "null"))

    # NEW: Parse boolean flag for circuit rendering
    show_circuit_flag = os.environ.get("VIEWER_SHOW_CIRCUIT", "0")
    show_circuit = True if show_circuit_flag == "1" else False

except Exception as e:
    print(f"Error parsing environment variables. Details: {e}")
    num_qubits = 3
    initial_state = None
    show_circuit = False

viewer = InteractiveDCNViewer(num_qubits=num_qubits, initial_state=initial_state)
viewer.display(show_circuit=show_circuit)
```
