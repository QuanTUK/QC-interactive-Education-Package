```python
import os
import json
from qc_education_package import InteractiveDCNViewer

# Safely extract parameters from the OS environment injected by the launcher
try:
    num_qubits = int(os.environ.get("VIEWER_QUBITS", 3))

    # Extract the initial state. 'null' is the JSON equivalent of Python's None.
    initial_state_raw = os.environ.get("VIEWER_INITIAL", "null")
    initial_state = json.loads(initial_state_raw)

except Exception as e:
    print(f"Error parsing environment variables. Defaulting to 3-qubit sandbox. Details: {e}")
    num_qubits = 3
    initial_state = None

# Initialize the interactive viewer
viewer = InteractiveDCNViewer(
    num_qubits=num_qubits,
    initial_state=initial_state
)

viewer.display()
```
