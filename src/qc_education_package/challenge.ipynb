```python
import os
import json
from qc_education_package import ChallengeDCNViewer

try:
    num_qubits = int(os.environ.get("CHALLENGE_QUBITS", 1))
    initial_state = json.loads(os.environ.get("CHALLENGE_INITIAL", "[1, 0]"))
    target_state = json.loads(os.environ.get("CHALLENGE_TARGET", "[1, -1]"))

    # NEW: Parse boolean flag for circuit rendering
    show_circuit_flag = os.environ.get("CHALLENGE_SHOW_CIRCUIT", "0")
    show_circuit = True if show_circuit_flag == "1" else False

except Exception as e:
    print(f"Error parsing environment variables. Details: {e}")
    num_qubits = 1
    initial_state = [1, 0]
    target_state = [1, -1]
    show_circuit = False

viewer = ChallengeDCNViewer(num_qubits=num_qubits, initial_state=initial_state, target_state=target_state)
viewer.display(show_circuit=show_circuit)
```
