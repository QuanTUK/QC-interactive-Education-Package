```python
import os
import json
from qc_education_package import ChallengeDCNViewer

# 1. Safely extract parameters from the OS environment injected by the launcher
try:
    num_qubits = int(os.environ.get("CHALLENGE_QUBITS", 1))
    initial_state = json.loads(os.environ.get("CHALLENGE_INITIAL", "[1, 0]"))
    target_state = json.loads(os.environ.get("CHALLENGE_TARGET", "[1, -1]"))
except Exception as e:
    print(f"Error parsing environment variables. Defaulting to 1-qubit challenge. Details: {e}")
    num_qubits = 1
    initial_state = [1, 0]
    target_state = [1, -1]

# 2. Initialize the dynamic challenge viewer
viewer = ChallengeDCNViewer(
    num_qubits=num_qubits,
    initial_state=initial_state,
    target_state=target_state
)

viewer.display()
```
